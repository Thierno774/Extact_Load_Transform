# packages Loading

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd 
```

# Get The URl of the sites

```python
url = "https://www.brawlplanet.com/brawlers"
page = requests.get(url)
```

```python
soup = bs(page.content, "html.parser")
```

```python
# print(soup.prettify())
```

# Find all the tables in the soup

```python
tables = soup.find_all( "tbody", attrs= {"divide-y divide-gray-200 dark:divide-gray-800"})
table_0 = tables[0]
brawler_percentage_list = []

for row in table_0.find_all("tr"):
    brawler_name = row.find_all("td")[0].text
    win_rate = row.find_all("td")[1].text
    use_rate = row.find_all("td")[2].text
    win_rate = row.find_all("td")[3].text

    percentage_per_brawler = []
    percentage_per_brawler.extend((brawler_name,win_rate,use_rate,win_rate ))
    brawler_percentage_list.append(percentage_per_brawler)
    #print(brawler_name)
    #win_rate = row.find_al
    #brawler_percentage_table.append(row)
```

```python
# Create a dataframe 
data = pd.DataFrame(brawler_percentage_list, 
                    columns=["Brawler name",
                            "Wine Rate (%)", 
                            "Use rate (%)", "Starplayer Rate (%)"])


data.head()
```

```python
table_1 = tables[1]
brawler_info_list = []

for row in table_1.find_all("tr"):
    brawler_name = row.find_all("td")[0].text
    health= row.find_all("td")[1].text
    attack = row.find_all("td")[2].text
    attack_range = row.find_all("td")[3].text
    reload_speed = row.find_all("td")[4].text 
    movement = row.find_all("td")[5].text

    info_per_brawler = []
    info_per_brawler.extend((brawler_name,
                             health,attack,attack_range, reload_speed, movement ))
    brawler_info_list.append(info_per_brawler)
## Create a dataframe 
dataset_info = pd.DataFrame(brawler_info_list, columns = ["Brawler name",
                                                          "Level 1 Health", 
                                                          "Level 1 Damage",
                                                          "Attack Range",
                                                          "Reload Speed",
                                                          "Movement"])

dataset_info.head()
```

# Merge Dataset

```python
data_all = pd.merge(data, dataset_info, on= "Brawler name" )
data_all.head()
```

# Save data

```python
data_all.to_csv("Brawler_name.csv")
```
